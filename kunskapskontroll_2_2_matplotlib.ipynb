```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The explicit object-oriented interface (or "Axes" interface) is the recommended approach as this gives you more options and is more customizable.
The other approach is called the implicit interface (or "pyplot" interface) and is faster and useful for quick and easy plots where the details are not as important. 

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is the whole figure and can contain one or more plots and keeps track of the elements in the plot, axes, artist, titles, figure legens, colors etc. 
Axes is the data area that is being plotted in the figure and contains the data being plotted.
Axis is the x- and y-axis within each axes
Artist is everything you see in the figure. When the figure is generated all the Artists are drawn to the canvas. 

### When plotting in Matplotlib, what is the expected input data type?

Numpy.array

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
# create a numpy.array with 100 equally distributed arguments between -4 and 4

y = x**2
# y is calculated as x squared

fig, ax = plt.subplots()
# create a figure with the axes for the plot

ax.plot(x, y)
# create a plots using the x arguments for the x-axis and the y arguments for the y-axis

ax.set_title('y = x^2')
# add the title "y = x^2" to the plot

ax.set_xlabel('x')
# add the axis-label "x" for the x-axis

ax.set_ylabel('y')
# add the axis-label "y" for the y-axis

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10,4), layout='constrained')
# fig, axs = plt.subplots(nrows, ncols)

axs[0].scatter(random_data_x,random_data_y)
axs[0].grid(True, linestyle='-', color='gray', alpha=0.5)
axs[0].set_axisbelow(True)
axs[0].set_title("Scatter plot")
axs[1].bar(names, values, color=["green", "blue", "red", "orange", "purple"])
axs[1].grid(True, axis='y', linestyle='-', color='gray', alpha=0.5)
axs[1].set_axisbelow(True)
axs[1].set_title("Bar plot")
plt.show()
```
